```python
from langchain.agents import create_agent 
from langchain.messages import SystemMessage, HumanMessage
from langchain_ollama import ChatOllama
from langgraph.checkpoint.memory import InMemorySaver
from tavily import TavilyClient
from dotenv import load_dotenv
from langchain.tools import tool


load_dotenv()
```

```python
from dotenv import load_dotenv
tavily_client = TavilyClient()
@tool("recette_ingredients")
def recette_ingredients(ingredients: str) -> str:
    """
    Generate a multiple recipes possible based on ingredients
    """
    print("tool recette_ingredients utilisé")
    # Logique pour générer la recette en fonction des ingrédients
    return f"Recette basée sur: {ingredients}. (logique à améliorer)" 

tavily_client = TavilyClient()

@tool('web_search')
def web_search(query: str) -> str:
    """Search the web for recipes or cooking information."""
    result = tavily_client.search(query)
    return str(result)

```

```python
# ======================
# LLM
# ======================
model = ChatOllama(model="llama3.2:3b")


# ======================
# AGENT create 
# ======================
agent = create_agent(
    model=model,
    tools=[web_search, recette_ingredients],
    checkpointer=InMemorySaver(),
)


config = {"configurable": {"thread_id": "1"}}

question = HumanMessage(content="J'ai du riz, poulet et tomates ?")
response = agent.invoke(
    {"messages": [question]},
    config=config
)
print(response['messages'][-1].content)
```
